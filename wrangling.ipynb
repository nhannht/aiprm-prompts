```python
import pandas
import glob
import re
```

```python
all_files = glob.glob("prompts/**/*.txt", recursive=True)
all_files
```

```python

patternTeaser = re.compile(r'Teaser:((.*\n)*?.*?)(\w+):', re.DOTALL)
patternAuthorName = re.compile(r'AuthorName: (.*)')
patternAuthorURL = re.compile(r'AuthorURL: (.*)')
patternTitle = re.compile(r'Title: (.*)')
patterCategory = re.compile(r'Category: (.*)')
patternPrompt = re.compile(r'Prompt:\n(.*$)', re.DOTALL)
patternCreationTime = re.compile(r'CreationTime: (.*)')
patternID = re.compile(r'ID: (.*)')
patternPromptHint = re.compile(r'PromptHint:((.*\n)*?.*?)(\w+):', re.DOTALL)
patternPromptPackageID = re.compile(r'PromptPackageID: (.*)')
```

```python
def convert_to_row (file):
    f = open(file, "r").read()
    authorName = patternAuthorName.search(f).group(1)
    authorURL = patternAuthorURL.search(f).group(1)
    title = patternTitle.search(f).group(1)
    category = patterCategory.search(f).group(1)
    prompt = patternPrompt.search(f).group(1)
    creationTime = patternCreationTime.search(f).group(1)
    id = patternID.search(f).group(1)
    promptHint = patternPromptHint.search(f).group(1)
    promptPackageID = patternPromptPackageID.search(f).group(1)
    teaser = patternTeaser.search(f).group(1)
    return {
        "authorName": authorName.strip(),
        "authorURL": authorURL.strip(),
        "title": title.strip(), 
        "category":  re.split("\s+",category.strip()),
        "prompt": prompt.strip(),
        "creationTime": creationTime.strip(),
        "id": id.strip(),
        "promptHint": promptHint.strip(),
        "promptPackageID": promptPackageID.strip(),
        "teaser": teaser.strip()
    }
    
```

```python
first_row = convert_to_row(all_files[0])
first_row
```

```python
df = pandas.DataFrame([convert_to_row(f) for f in all_files])    
```
